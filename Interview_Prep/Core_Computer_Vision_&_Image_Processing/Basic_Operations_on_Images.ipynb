```python
import numpy as np
import cv2 as cv
import matplotlib.pyplot as plt

img = cv.imread('../1540.jpg')
img = cv.cvtColor(img, cv.COLOR_BGR2RGB)
assert img is not None, "file could not be read, check with os.path.exists()"
```

```python
px = img[100, 100]
px
```

```python
img[100,100] = [255,255,255]
plt.imshow(img)
```

```python
face = img[20:450, 330:800]
plt.imshow(face)
```

```python
face_rs = cv.resize(face, (120, 120))
img[20:140, 530:650] = face_rs
plt.imshow(img)
```

```python
b,g,r = cv.split(img)
b_ = np.zeros(b.shape, dtype=b.dtype)
img_garbage = cv.merge((b_,g,r))
plt.imshow(img_garbage)
```
