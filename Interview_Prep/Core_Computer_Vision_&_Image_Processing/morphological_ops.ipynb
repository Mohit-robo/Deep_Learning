```python
import numpy as np
import cv2 
import matplotlib.pyplot as plt
```

```python
image = cv2.imread('../lether_stitches.jpeg')
image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
plt.imshow(image, cmap='gray')
```

```python
_, binary = cv2.threshold(image, 80, 250, cv2.THRESH_BINARY_INV)
plt.imshow(binary, cmap='gray')
```

```python
kernel = cv2.getStructuringElement(cv2.MORPH_RECT, (5,5))
dilation = cv2.dilate(binary, kernel, iterations=1)
plt.imshow(dilation, cmap='gray')
```

```python
erosion = cv2.erode(binary, kernel=kernel, iterations=2)
plt.imshow(erosion, cmap='gray')
```

```python
opening = cv2.morphologyEx(binary, cv2.MORPH_OPEN, kernel=kernel, iterations=1)
plt.imshow(opening, cmap='gray')
```

```python
close = cv2.morphologyEx(binary, cv2.MORPH_CLOSE, kernel=kernel, iterations=1)
plt.imshow(opening, cmap='gray')
```
